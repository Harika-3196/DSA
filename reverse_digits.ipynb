```
def reverse_digits(n):
    reversed_number=0
    while n>0:
        last_digit=n%10
        reversed_number=(reversed_number*10)+last_digit
        n=n//10
    return reversed_number
```

# Time complexity
in the while loop we divide N by 10 until 0 which it become 1og10N iterations.In each iteration og while loop we perform constant time operation like modulus so it is 1og10N+1

# space complexity
o(1) is the only constant amount of additional memory for reversed number regardless of size of input number
